```python
use HTTP::UserAgent;
use DOM::HTML;
use Time::Duration;

my $url = 'https://github.com/Raku/problem-solving/issues/';

my $ua = HTTP::UserAgent.new;
my $res = $ua.get($url);

unless $res.is-success {
    die "Failed to fetch $url: {$res.status-line}\n";
}

my $dom = DOM::HTML.parse($res.content);

# Get current time in UTC
my $now = DateTime.now;

# Find all issues on page
# On GitHub issues page, each issue is in an <a> with data-hovercard-type="issue"
# But better to select items under div with role="row" or under 'js-issue-row' class
# Each issue row has 'js-issue-row' class and an id like "issue_123"
# The title link is inside the row, usually under a <a> with class "Link--primary"

my @issues;

for $dom.query('.js-issue-row') -> $issue-row {
# Issue title link
my $title-a = $issue-row.query('a.Link--primary');
next unless $title-a;

my $issue-title = $title-a.text.trim;
my $issue-link  = 'https://github.com' ~ $title-a<href>;

# The relative time is inside <relative-time datetime="2024-06-16T12:34:56Z">
my $reltime = $issue-row.query('relative-time');
next unless $reltime;

my $datetime-attr = $reltime<datetime>;
next unless $datetime-attr;

my $issue-date = DateTime.new($datetime-attr);

# Calculate if within last 7 days
if $issue-date >= $now - 7*24*60*60 {
@issues.push: { title => $issue-title, url => $issue-link }
}
}

say '<ul>';
for @issues -> $issue {
say qq:to/HTML/;
<li><a href="{$issue<url>}">{$issue<title>}</a></li>
HTML
}
say '</ul>';
```
